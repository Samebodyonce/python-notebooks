```python
import numpy as np
%matplotlib inline
from matplotlib import pylab as plt
from scipy import linalg
import math
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
x = np.arange(1, 16)
plt.plot(x, f(x))
```

```python

A1 = np.array([[1, 1], [1, 15]])
b1 = np.array([f(1), f(15)])
ans1 = linalg.solve(A1, b1)
A2 = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]])
b2 = np.array([f(1), f(8), f(15)])
ans2 = linalg.solve(A2, b2)
A3 = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3],[1, 15, 15**2, 15**3]])
b3 = np.array([f(1), f(4), f(10), f(15)])
ans3 = linalg.solve(A2, b2)
A3 = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3],[1, 15, 15**2, 15**3]])
b3 = np.array([f(1), f(4), f(10), f(15)])
ans3 = linalg.solve(A3, b3)
print(A3, b3, ans3)
def f1(x):
    return ans1[0]+ans1[1]*x
def f2(x):
    return ans2[0]+ans2[1]*x+ans2[2]*x**2
def f3(x):
    return ans3[0]+ans3[1]*x+ans3[2]*x**2+ans3[3]*x**3
plt.plot(x, f(x))
plt.plot(x, f1(x))
plt.plot(x, f2(x))
plt.plot(x, f3(x))
```
